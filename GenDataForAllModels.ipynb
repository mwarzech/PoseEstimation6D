```python
def find_datapaths_for_models(models_names):
    paths = {}
    for data_dir in tqdm(data_dirs):
        txt_files = [file for file in os.listdir(data_dir) if file.endswith(".txt")]
        for txt_file in txt_files:
            with open(os.path.join(data_dir, txt_file)) as f:
                file_content = f.read()
                for model_name in models_names:
                    if model_name in file_content:
                        if model_name not in paths:
                            paths[model_name] = []
                        paths[model_name].append(os.path.join(data_dir, txt_file).replace("\\", "/"))
    return paths
```

```python
def find_model_label_index(model_name):
    img_path = models_img_paths[model_name][0]
    num = 0
    with open(img_path.replace('color.png', 'box.txt')) as f:
        for i, line in enumerate(f):
            if model_name in line:
                num = i
                break

    return scipy.io.loadmat(img_path.replace('color.png', 'meta.mat'))['cls_indexes'][num][0]
```

### Objects models and poses

```python
model_3d_paths = {}
poses_data_paths = {}

for model_name in models_names:
    poses_data_paths[model_name] = 'YCB_Video_Dataset/poses/' + [x for x in os.listdir('YCB_Video_Dataset/poses/') if model_name in x][0]
    model_3d_paths[model_name] = 'YCB_Video_Dataset/models/' + [x for x in os.listdir('YCB_Video_Dataset/models/') if model_name in x][0] + '/points.xyz'
```

```python
models_poses = {}
models_points = {}

for model_name in models_names:
    models_poses[model_name] = np.array(pd.read_csv(poses_data_paths[model_name], sep = ' ', header=None))
    models_points[model_name] = np.array(pd.read_csv(model_3d_paths[model_name], sep = ' ', header=None))
```

```python
def pose_to_RtK(pose, path):
    t = pose[4:,None]
    quat = -np.array([*pose[1:4], pose[0]])
    R = np.array(Rotation.from_quat(quat).as_matrix())
    K = K_dict[path.replace('color.png', 'meta.mat')]
    return {'R': R, 't': t, 'K': K}
```

```python
models_poses_dict = {}

models_names = [model_name for model_name in models_names if len(models_img_paths[model_name]) == len(models_poses[model_name])]
for model_name in models_names:
    models_poses_dict[model_name] = {models_img_paths[model_name][i]: pose_to_RtK(models_poses[model_name][i], models_img_paths[model_name][i]) for i in tqdm(range(len(models_img_paths[model_name])))}
```

### Prepare data format for Hourglass model

```python
def is_in_bounds(pos, img):
    return (int(pos[1]) > 0 and int(pos[1]) < img.shape[0] and \
            int(pos[0]) > 0 and int(pos[0]) < img.shape[1])
```

```python
def get_label_on_image(pos, img):
    if not is_in_bounds(pos, img):
        return 0
    return int(img[int(pos[1]), int(pos[0])] * 255)
```

```python
def load_data(dataset, model_name, path):
    RtK = models_poses_dict[model_name][path]
    R = RtK['R']
    t = RtK['t']
    K = RtK['K']
    img = mpimg.imread(path)
    verts = models_points[model_name]

    verts_2d = np.matmul(K, np.matmul(R, verts.T) + t).T
    verts_2d = verts_2d[:,:2] / verts_2d[:,2,None]

    keypoints_2d = verts_2d[::100]
    
    img_labels = mpimg.imread(path.replace("color", "label"))
    target_weight = []
    for i in range(len(keypoints_2d)):
        weight = 0
        if  models_labels_dict[model_name] == get_label_on_image(keypoints_2d[i], img_labels) or \
            models_labels_dict[model_name] == get_label_on_image(keypoints_2d[i] + np.array([0,2]), img_labels) or\
            models_labels_dict[model_name] == get_label_on_image(keypoints_2d[i] + np.array([0,-2]), img_labels) or\
            models_labels_dict[model_name] == get_label_on_image(keypoints_2d[i] + np.array([2,2]), img_labels) or\
            models_labels_dict[model_name] == get_label_on_image(keypoints_2d[i] + np.array([-2,2]), img_labels) or\
            models_labels_dict[model_name] == get_label_on_image(keypoints_2d[i] + np.array([2,-2]), img_labels) or\
            models_labels_dict[model_name] == get_label_on_image(keypoints_2d[i] + np.array([-2,-2]), img_labels) or\
            models_labels_dict[model_name] == get_label_on_image(keypoints_2d[i] + np.array([2,0]), img_labels) or\
            models_labels_dict[model_name] == get_label_on_image(keypoints_2d[i] + np.array([-2,0]), img_labels):
                weight = 1
        target_weight.append(weight)
    
    x_max = np.max(verts_2d[:,0])
    x_min = np.min(verts_2d[:,0])
    y_max = np.max(verts_2d[:,1])
    y_min = np.min(verts_2d[:,1])
    scale = np.max([x_max - x_min, y_max - y_min])/(200-10)
    center = [np.mean([x_max, x_min]), np.mean([y_max, y_min])]
    
    dataset['imgname'].append(path)
    dataset['part'].append(keypoints_2d)
    dataset['center'].append(center)
    dataset['scale'].append(scale)
    dataset['target_weight'].append(np.array(target_weight))
```

```python
def load_data_for_model(model_name):
    dataset_train = {}
    dataset_train['imgname'] = []
    dataset_train['part'] = []
    dataset_train['center'] = []
    dataset_train['scale'] = []
    dataset_train['target_weight'] = []

    dataset_test = {}
    dataset_test['imgname'] = []
    dataset_test['part'] = []
    dataset_test['center'] = []
    dataset_test['scale'] = []
    dataset_test['target_weight'] = []

    for path in tqdm(models_img_paths[model_name]):
        load_data(dataset_test if is_path_test(path, test_set) else dataset_train, model_name, path)
        
    return dataset_train, dataset_test
```

```python
def generate_train_and_valid_data_for_model(model_name):
    dataset_train, dataset_test = load_data_for_model(model_name)
    np.savez(model_name + '_valid_data.npz', **dataset_test)
    np.savez(model_name + '_train_data.npz', **dataset_train)
```

```python
def generate_metadata_for_model(model_name):
    metadata = {}
    verts = models_points[model_name]
    metadata['keypoints_3d'] = verts[::100]
    metadata['verts'] = verts
    metadata['poses'] = models_poses_dict[model_name]
    np.savez(model_name + '_meta_data.npz', **metadata)
```

```python
def calculate_diameter(model_name):
    verts = models_points[model_name]
    diameter = 0

    for v1 in tqdm(verts):
        for v2 in verts:
            dist = np.linalg.norm(v1 - v2)
            if dist > diameter:
                diameter = dist
    return diameter
```

```python
models_diameters = {}

for model_name in models_names:
    models_diameters[model_name] = calculate_diameter(model_name)
```
